<h1 style = "font-size:3rem;color:red;" >IF STATEMENT</h1>

The if-else statement is used to execute both the true part and the false part of a given condition. If the condition is true, the if block code is executed and if the condition is false, the else block code is executed.

An "if statement" is written by using the if keyword.

```python
__syntax:__
    
if condition :
    operation
```

```python
#Example
a = 33
b = 200
if b > a:
  print("b is greater than a")
```

If you use wrong syntax or indentation you will get the error in the execution of the program.

```python
a = 33
b = 200
if b > a:
print("b is greater than a") # you will get an error 
```

<h1 style = "font-size:3rem;color:red;" >ELSE STATEMENT</h1>

The else keyword catches anything which isn't caught by the preceding conditions.

```python
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
if a>b:
    print("HELLO BUDDY")
else:
    print("BYE BUDDY")
```

<h1 style = "font-size:3rem;color:red;" >ELIF STATEMENT</h1>

The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition".

```python
a=int(input("Enter the first number:"))
b=int(input("Enter the second number:"))
if a>b:
    print("HELLO BUDDY")
elif a==b:
    print("WE ARE THE SAME BRO")
else:
    print("BYE BUDDY")
```

<h1 style = "font-size:3rem;color:red;" >AND STATEMENT</h1>

The and keyword is a logical operator, and is used to combine conditional statements:

__And is used when the both condition is true.__

```python
a=3
if a<33 and a<21:
    print("a is bigger now")
```

<h1 style = "font-size:3rem;color:red;" >OR STATEMENT</h1>

or statement is used when atleast one condition is true.

```python
a=23
if a<33 or a<21:
    print("a is bigger")
```

<h1 style = "font-size:3rem;color:red;" >NESTED IF STATEMENT</h1>

When the two or more if statement is use in between the if statement then it is called the Nested If.

```python
x =int(input("Enter the number:"))
if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
else:
    print("smaller then 10")  # execute when the input number is smaller than the 10
```

<h1 style = "font-size:3rem;color:red;" >PASS STATEMENT</h1>

if statements cannot be empty, but if you for some reason have an if statement with no content, put in the pass statement to avoid getting an error.

```python
a = 33
b = 200

if b > a:
  pass
```
